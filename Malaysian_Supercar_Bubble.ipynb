```python
import matplotlib.pyplot as plt
import pandas as pd

# --- DATA ---
data = {
    'Year': [2020, 2021, 2022, 2023, 2024, 2025],
    'Luxury_EV_Sales_Growth': [5, 8, 35, 45, 60, 48],
    'Household_Debt_to_GDP': [82.0, 84.5, 86.0, 84.2, 85.1, 86.5],
    'Young_Bankruptcy_Rate': [18.2, 19.5, 20.1, 21.5, 23.6, 25.1],
    'Vehicle_Loan_Default_Rate': [2.1, 2.3, 2.8, 3.2, 4.5, 5.1]
}
df = pd.DataFrame(data)

# --- DARK MODE SETUP ---
# Set the overall style params manually for total control
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'

# Create Figure with Black Background
fig, ax1 = plt.subplots(figsize=(12, 7), facecolor='black')
ax1.set_facecolor('black')

# 1. The "Flex" (Sales Growth) - Bars
# Used a brighter blue/cyan to pop against black
bars = ax1.bar(df['Year'], df['Luxury_EV_Sales_Growth'], color='#00d2ff', alpha=0.7, label='Luxury Sales Growth (%)')
ax1.set_ylabel('Sales Growth (%)', color='#00d2ff', fontsize=11, fontweight='bold')
ax1.tick_params(axis='y', labelcolor='#00d2ff')
ax1.set_xlabel('Year', fontsize=12, color='white')

# 2. The "Pain" (Default Rates) - Line
ax2 = ax1.twinx()
# Bright Neon Red for visibility
line = ax2.plot(df['Year'], df['Vehicle_Loan_Default_Rate'], color='#ff4d4d', marker='o', linewidth=3, label='Car Loan Default Rate (%)')
ax2.set_ylabel('Loan Default Rate (%)', color='#ff4d4d', fontsize=11, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='#ff4d4d')

# Remove the white box around the second axis to blend it in
ax2.set_facecolor('black') # Transparent/Black

# --- CLEANUP & STYLING ---

# Title
plt.title('THE MALAYSIAN "SUPERCAR BUBBLE": FLEX VS REALITY (2020-2025)', fontsize=14, fontweight='bold', pad=40, color='white')

# Legend: White text, no background box
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
legend = ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2, frameon=False, fontsize=10)
plt.setp(legend.get_texts(), color='white')

# Annotation: Dark Grey Box with Bright Text
plt.annotate(
    'THE "SAMBUNG BAYAR" SPIKE\n(Tax Exemption Ends + Rates Rise)',
    xy=(2024, 4.5),
    xytext=(2021, 4.2),
    arrowprops=dict(facecolor='white', arrowstyle='->', lw=1.5, color='white'), # White Arrow
    fontsize=9,
    backgroundcolor='#333333',  # Dark Grey box instead of white
    color='#ff4d4d',            # Red Text to match the line
    fontweight='bold',
    ha='center',
    bbox=dict(boxstyle="square,pad=0.3", fc="#333333", ec="none") # Clean box
)

# Grid: Low opacity white lines
ax1.grid(True, linestyle='--', alpha=0.2, color='white')
ax1.set_axisbelow(True) # Put grid behind bars

# Spines (Borders): Make them white
for ax in [ax1, ax2]:
    ax.spines['bottom'].set_color('white')
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_color('white')
    ax.spines['right'].set_color('white')

plt.tight_layout()

# Save/Show
plt.show()
```

```python
import pandas as pd
import random
from IPython.display import display

# --- 1. CONFIGURATION (The "Menu") ---
MODELS = {
    "BMW M5 F90": {"Base_Price": 450000, "Beta": 3.0},
    "Honda Civic Type R": {"Base_Price": 240000, "Beta": 1.0},
    "Mercedes C200": {"Base_Price": 180000, "Beta": 1.2},
    "McLaren 720S": {"Base_Price": 950000, "Beta": 4.0},
    "Ferrari 488 GTB": {"Base_Price": 1100000, "Beta": 1.5},
    "Lamborghini Huracan": {"Base_Price": 980000, "Beta": 1.8},
    "Bentley Continental GT": {"Base_Price": 550000, "Beta": 3.5},
    "Rolls Royce Ghost": {"Base_Price": 850000, "Beta": 2.5},
    "Porsche 911 GT3": {"Base_Price": 1300000, "Beta": 1.1}
}

# --- 2. SCENARIO ENGINE ---
SCENARIOS = [
    {"type": "Sambung Bayar", "desc_pool": ["monthly sambung", "continue loan kedai"], "discount": 0.1, "risk_add": 20},
    {"type": "Lari Bank", "desc_pool": ["lari bank", "no document", "kereta piang"], "discount": 0.4, "risk_add": 30},
    {"type": "Migration", "desc_pool": ["owner migration", "moving overseas"], "discount": 0.15, "risk_add": 5},
    {"type": "Divorce", "desc_pool": ["divorce settlement", "liquidate asset"], "discount": 0.2, "risk_add": 5},
    {"type": "Retail", "desc_pool": ["excellent condition", "low mileage"], "discount": -0.05, "risk_add": 0}
]

def generate_market_feed(n=50):
    data = []
    print(f">>> CONNECTING TO SHADOW MARKET FEEDS... SCANNED {n} LISTINGS.")

    for _ in range(n):
        model_name, specs = random.choice(list(MODELS.items()))
        scenario = random.choices(SCENARIOS, weights=[20, 10, 15, 5, 50], k=1)[0]

        fmv = specs['Base_Price']
        ask_price = fmv * (1 - scenario['discount']) * random.uniform(0.95, 1.05)

        if scenario['type'] in ["Sambung Bayar"]:
            monthly = (ask_price * 0.9) / 60
            upfront = ask_price * 0.1
            months_left = random.randint(48, 108)
            structure = "Financed (Sub-Prime)"
            total_cost = upfront + (monthly * months_left)
        else:
            structure = "Cash / Settlement"
            total_cost = ask_price

        risk_score = scenario['risk_add']
        if specs['Beta'] > 2.5: risk_score += 5

        margin = ((fmv - total_cost) / fmv) * 100

        desc = f"{scenario['type']} - {random.choice(scenario['desc_pool'])}"
        if "lari" in desc or "no doc" in desc: verdict = " REGULATORY NUKE"
        elif margin < 0: verdict = " UNDERWATER (TRAP)"
        elif margin > 15 and risk_score < 15: verdict = " PRIME ALPHA"
        elif margin > 25: verdict = " HIGH RISK GAMBLE"
        else: verdict = "WAIT / IGNORE"

        data.append({
            "Asset": model_name,
            "Deal Structure": structure,
            "FMV (Est)": fmv,
            "Ask Price/Cost": total_cost,
            "Spread %": margin,
            "Verdict": verdict,
            "Risk Flags": scenario['type']
        })

    return pd.DataFrame(data)

# --- EXECUTION ---
df_feed = generate_market_feed(40)
display_df = df_feed[df_feed['Verdict'] != "WAIT / IGNORE"].sort_values(by="Spread %", ascending=False).head(12)

# --- 3. STYLING ENGINE (SOLID CHIPS) ---
def highlight_verdict(val):
    # 1. PRIME (Solid Neon Green Block)
    if "PRIME" in val:
        return 'background-color: #00ff00; color: black; font-weight: bold; border-radius: 4px;'

    # 2. UNDERWATER (Solid Red Block)
    if "UNDERWATER" in val:
        return 'background-color: #ff0000; color: white; font-weight: bold; border-radius: 4px;'

    # 3. NUKE (Solid Orange Block)
    if "NUKE" in val:
        return 'background-color: #FFA500; color: black; font-weight: bold; border-radius: 4px;'

    # 4. HIGH RISK (Solid Yellow Block)
    if "HIGH RISK" in val:
        return 'background-color: #FFFF00; color: black; font-weight: bold; border-radius: 4px;'

    return ''

print("\n>>> [QUANT TERMINAL] LIVE MARKET OPPORTUNITY FEED:")

# Apply Styling
# Note: We set the global properties FIRST, then apply the map to override specific cells.
styled_df = display_df.style.set_properties(**{
        'background-color': 'black',
        'color': 'white',
        'border-color': '#444'
    })\
    .map(highlight_verdict, subset=['Verdict'])\
    .format({"FMV (Est)": "{:,.0f}", "Ask Price/Cost": "{:,.0f}", "Spread %": "{:.1f}%"})\
    .hide(axis="index")\
    .set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#333'), ('color', 'white'), ('font-weight', 'bold'), ('font-size', '14px')]},
        {'selector': 'td', 'props': [('padding', '8px')]}
    ])

display(styled_df)
```
